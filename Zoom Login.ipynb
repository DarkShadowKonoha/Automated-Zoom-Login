```python
option = Options()

option.add_argument("--disable-infobars")      #Disable Inforbars
option.add_argument("start-maximized")         #Start Window Maximized
option.add_argument("--disable-extensions")    #Disables Browser Extensions for this session


# Setting mic to mute and camera to off
# 0 for mute/off and 1 for unmute/on
option.add_experimental_option("prefs", {
 "profile.default_content_setting_values.notifications":       1,
 "profile.default_content_setting_values.media_stream_mic":    1,
 "profile.default_content_setting_values.media_stream_camera": 1,
 })
```

```python
# passing the above options to our webdriver
driver = webdriver.Chrome(options=option)
```
